```python
# Passo 1: Importar a base de dados

import pandas as pd

tabela = pd.read_csv("telecom_users.csv")
    
# Passo 2: Visualizar a base de dados
# Entender as informações que você tem disponível
# Descobrir as cagadas da base de dados

# excluir coluna inútil
# axis = 0 -> eixo da linha
# axis = 1 -> eixo da coluna

tabela = tabela.drop("Unnamed: 0", axis=1)
display(tabela)
```

```python
# Passo 3: Tratamento de dados (resolver as cagadas da base de dados)
# Informações do tipo correto - ajustar o TotalGasto - transformar para númerico, e forçar a forçar número os erros

tabela["TotalGasto"] = pd.to_numeric(tabela["TotalGasto"], errors="coerce")

# Informação inútil

# colunas vazias -> excluir
tabela = tabela.dropna(how="all", axis=1)

# Linhas que tem alguma informação vazia (como são poucas remover)
tabela = tabela.dropna(how="any", axis=0)


print(tabela.info())
```

```python
# Passo 4: Análise inicial dos dados
# Como estão os cancelamentos? 26%

print(tabela["Churn"].value_counts())

print(tabela["Churn"].value_counts(normalize=True).map("{:.1%}".format))
```

```python
# Passo 5: Descobrir os motivos do cancelamento
import plotly.express as px

# etapa 1: cria o gráfico
grafico = px.histogram(tabela, x="MesesComoCliente", color="Churn")

# etapa 2: exibe o gráfico
grafico.show()

grafico = px.histogram(tabela, x="Aposentado", color="Churn")
grafico.show()

grafico = px.histogram(tabela, x="Genero", color="Churn")
grafico.show()

grafico = px.histogram(tabela, x="Casado", color="Churn")
grafico.show()
```
